# Преобразование данных

1. Преобразование признака **"Образование и ВУЗ"**. 
Текущий формат: **<Уровень образования год выпуска ВУЗ специальность...>**. 
Требуется выделить только уровень образования, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Вопрос: Сколько соискателей имеет средний уровень образования?

```python
hh_data['Образование'] = hh_data['Образование и ВУЗ'].apply(
    lambda x: (str(x)).split(' ')[0]+' '+(str(x)).split(' ')[1]\
    if (str(x)).split(' ')[1] != 'образование' else (str(x)).split(' ')[0]) # оставляем только уровень образования
hh_data.drop('Образование и ВУЗ', axis = 1, inplace=True) # удаляем старый признак

print(f"Средний уровень образования имеет {hh_data['Образование'].value_counts()['Среднее']} соискателей")
```

2. Преобразование признака **"Пол, возраст"**. 
Сейчас он представлен в формате **<Пол , возраст , дата рождения >**. 
Нужно создать два новых признака **"Пол"** и **"Возраст"** и удалить признак **"Пол, возраст"** из таблицы.
Вопросы:
Сколько процентов женских резюме представлено в наших данных? Ответ округлите до сотых.
Чему равен средний возраст соискателей? Ответ округлите до десятых.

```python
# заменяем значения "Мужчина" и "Женщина" на "М" и "Ж"
hh_data['Пол'] = hh_data['Пол, возраст'].apply(
    lambda x: 'М'if (str(x)).split(' ')[0] == 'Мужчина' else 'Ж' 
) 
hh_data['Пол'] = hh_data['Пол'].astype('category') # меняем тип признака на category 

# оставляем только число возраста int 
hh_data['Возраст'] = hh_data['Пол, возраст'].apply(
    lambda x: (str(x)).split(' ')[3]
)
hh_data['Возраст'] = hh_data['Возраст'].astype('int')
hh_data.drop('Пол, возраст', axis=1, inplace=True) # удаляем старый признак


print(f"Среди соискателей представленно {round(hh_data['Пол'].value_counts(normalize=True).loc['Ж']*100, 2)}% резюме женщин")
print(f"Средний возраст соискателей составляет {round(hh_data['Возраст'].mean(), 2)} года")
```

3. Преобразование признака **"Опыт работы"**. 
Его текущий формат: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Условия задачи:
* Во-первых, в данном признаке есть пропуски. Если мы встречаем пропуск, оставляем его как есть.
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN.
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях).
Удалить столбец **"Опыт работы"** из таблицы.

Вопросы:

Чему равен медианный опыт работы (в месяцах) в нашей таблице?

```python
def work_xp(arg):
    """Функция преобразовывает значения столбца "Опыт работы" в количество месяцев

    Args:
        arg ([string]): строка с данными об опыте работы 

    Returns:
        [string]: количество месяцев рабочего опыта
    """
    # объявляем переменные для лет и месяцев
    months = 0
    years = 0
    if arg is np.nan or arg == 'Не указано': return None # оставляем Nan и заменяем "Не указано" на Nan 
    
    arg = list(arg.split())
    if arg[3] in ['месяц', 'месяца', 'месяцев']: # проверяем, указан ли год,
        months = int(arg[2]) # если нет, оставляем количество месяцев
    if arg[3] in ['год', 'года', 'лет']:
        years = int(arg[2])*12 # если указан год, умножаем его на 12
    if arg[5] in ['месяц', 'месяца', 'месяцев']:
        months = int(arg[4]) # проверяем указаны ли месяцы после года
    return years+months
# создаём новый признак
hh_data['Опыт работы (месяц)'] = hh_data['Опыт работы'].apply(work_xp)
hh_data.drop('Опыт работы', axis=1, inplace=True) # удаляем столбец

print(f"Медианный опыт работы составляет {hh_data['Опыт работы (месяц)'].median()} месяцев")
```

4. Преобразование признака "Город, переезд, командировки". 
Информация в нем представлена в следующем виде: **<Город , (метро) , готовность к переезду (города для переезда) , готовность к командировкам>**. В скобках указаны необязательные параметры строки. 

Создадать отдельные признаки **"Город"**, **"Готовность к переезду"**, **"Готовность к командировкам"**. 
При этом важно учесть:
* Признак **"Город"** должен содержать только 4 категории: "Москва", "Санкт-Петербург" и "город-миллионник" (их список ниже), остальные обозначить как "другие".

    Список городов-миллионников:
    
   <code>million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск','Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград']
    </code>
    Инфорация о метро, рядом с которым проживает соискатель нас не интересует.

* Признак **"Готовность к переезду"** должен иметь два возможных варианта: True или False.

* Признак **"Готовность к командировкам"** должен иметь два возможных варианта: True или False. 
    
    Еще один важный факт: при выгрузки данных у некоторых соискателей "потерялась" информация о готовности к командировкам. По умолчанию считать, что такие соискатели не готовы к командировкам.
    
Удалить столбец **"Город, переезд, командировки"** из таблицы.

Вопросы:
Сколько процентов соискателей живут в Санкт-Петербурге?

Сколько процентов соискателей готовы одновременно и к переездам, и к командировкам?


```python
# Функция для определения типа города
def city(arg):
    arg = list(arg.split())
    million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 
                      'Казань', 'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 
                      'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград' ]
    if arg[0] in million_cities:
        arg = 'город-миллионник'
    elif arg[0] == 'Москва':
        arg = arg[0]
    elif arg[0] in ['Санкт-Петербург', 'Спб', 'СПБ', 'СПб', 'Санкт Петербург']:
        arg = 'Санкт-Петербург'
    else:
        arg = 'другие'
    return arg

# Функция для определения готовности к переездам
def relocation(arg):
    arg = list(arg.split('пере'))
    if 'не' in list(arg[0].split()):
        return False
    else:
        return True
# Функция для определения гтовности к командировкам
def business_trip(arg):
    arg = list(arg.split('пере'))
    if 'не' in list(arg[1].split()) or 'командировкам' not in list(arg[1].split()):
        return False
    else:
        return True
# создание новых признаков    
hh_data['Город'] = hh_data['Город, переезд, командировки'].apply(city)
hh_data['Готовность к переезду'] = hh_data['Город, переезд, командировки'].apply(relocation)
hh_data['Готовность к командировкам'] = hh_data['Город, переезд, командировки'].apply(business_trip)
hh_data.drop(columns='Город, переезд, командировки', inplace=True) # удаление старого признака

print(f"В Санкт-Петербурге живут {round(hh_data['Город'].value_counts(normalize=True).loc['Санкт-Петербург']*100)}% соискателей")
reloc = hh_data['Готовность к переезду'] == True
trip = hh_data['Готовность к командировкам'] == True
print(f"Доля соискателей готовых к переезду и командировкам - {(round(hh_data[reloc & trip].count()[0]/hh_data.shape[0]*100))}% от общего числа") 
```

5. Преобразование признаков **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу.

Нужно создать признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Удалить признаки **"Занятость"** и **"График"**.

Вопросы:
Сколько людей ищут проектную работу и волонтёрство?

Сколько людей хотят работать вахтовым методом и с гибким графиком?

```python
# Виды занятости 
employment = ['полная занятость', 'частичная занятость', 'проектная работа', 'волонтерство', 'стажировка']
# Создание новых признаков
for el in employment:
    hh_data[el] = hh_data['Занятость'].apply(lambda x: el in x)
shedule = ['полный день', 'сменный график', 'гибкий график', 'удаленная работа', 'вахтовый метод']
for el in shedule:
    hh_data[el] = hh_data['График'].apply(lambda x: el in x)
    
hh_data.drop(columns='Занятость', inplace=True) # удаление старого признака
hh_data.drop(columns='График', inplace=True) # удаление старого признака

e1 = (hh_data[(hh_data['проектная работа'] == True) & (hh_data['волонтерство'] == True)].shape[0])
e2 = (hh_data[(hh_data['вахтовый метод'] == True) & (hh_data['гибкий график'] == True)].shape[0])
print(f"{e1} людей ищут проектную работу и волонтёрство")
print(f"{e2} людей хотят работать вахтовым методом и с гибким графиком")
```

6. Преобразование признака заработной платы **"ЗП"**. 
Помимо желаемой заработной платы соискатель указывает валюту, в которой он бы хотел ее получать.
Хотелось бы видеть заработную плату в единой валюте - в рублях. 

Обратимся к специальным интернет-ресурсам для получения данных о курсе в виде текстовых файлов. Например, MDF.RU, данный ресурс позволяет удобно экспортировать данные о курсах различных валют и акций за указанные периоды в виде csv файлов.


<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/table.jpg>

Нужно создать признак "ЗП (руб)" - заработная плата в рублях.

Удалить исходный столбец заработной платы "ЗП" и все промежуточные столбцы.

Вопрос:
Чему равна желаемая медианная заработная плата соискателей в нашей таблице?

Алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
many_table = pd.read_csv('data/ExchangeRates.csv', sep=',') # Считаем таблицу с данными по обмену валюты
many_table['date']=pd.to_datetime(many_table['date']) # переведем дату в формат datetime
many_table['date']=many_table['date'].dt.date

# также переведем дату в таблице с резюме и отбросим время
hh_data['Обновление резюме'] = pd.to_datetime(hh_data['Обновление резюме'])
hh_data['Обновление резюме'] = hh_data['Обновление резюме'].dt.date

# Функция для выделения числа в значении ЗП
def get_salary_num(arg):
    s = float(arg.split(' ')[0])
    return s
# Функция для переименования валюты
def rename_currency(arg):
    """Функция считывает наименование валюты и если она не соответствует международному стандарту возвращает стандартное наименование

    Args:
        arg ([string]):Строка содержащая наименование валюты

    Returns:
        [string]: Наименование валюты в стандартном виде
    """
    currency = {'USD': 'USD', 'KZT': 'KZT', 'грн': 'UAH', 'белруб': 'BYN', 'EUR': 'EUR', 'KGS': 'KGS', 'сум': 'UZS', 'AZN': 'AZN'}
    curr = arg.split(' ')[1].replace('.', '')
    if curr == 'руб':
        return 'RUB'
    else:
        return currency[curr]
# Применяем функции
hh_data['Сумма'] = hh_data['ЗП'].apply(get_salary_num)
hh_data['Валюта'] = hh_data['ЗП'].apply(rename_currency)    

# Объединим оба DataFrame, для чего используем метод merge
merged = hh_data.merge(
    many_table,
    how = 'left',
    left_on=['Валюта', 'Обновление резюме'],
    right_on=['currency','date']    
)

# Заполним столбец close и proportion значением 1.0 для рубля
merged['close'] = merged['close'].fillna(1)
merged['proportion'] = merged['proportion'].fillna(1)
# Создадим новый признак заработной платы
hh_data['ЗП (руб)'] =  merged['close'] * merged['Сумма']/ merged['proportion']
# Удалим ненужные признаки
hh_data.drop(labels=['ЗП', 'Сумма', 'Валюта'], axis=1, inplace=True)

print(f"Медианная желаемая ЗП = {round(hh_data['ЗП (руб)'].median()/1000)} тысяч рублей")
```

```python

fig, ax = plt.subplots(1, 1, figsize=(8, 4))
log_age = np.log(hh_data['Возраст'])# не прибавляем 0, так как у нас нет нулевых значений возраста
histplot = sns.histplot(log_age, bins=30, ax=ax)
# Построение линий
histplot.axvline(log_age.mean(), color='r', lw=2)
histplot.axvline((log_age.mean()+ 3 * log_age.std()), color='k', ls='--', lw=2) # пунктир
histplot.axvline((log_age.mean()- 3 * log_age.std()), color='k', ls='--', lw=2) # пунктир
histplot.set_title('Логарифмическое аспределение признака Возраст');
```

```python
# функция метода z-отклонения

def outliers_z_score_mod(data, feature, log_scale=False, left=3, right=3):
    """Функция очищает данные методом z-отклонения
    Args:
        data (_type_): данные на входе
        feature (_type_): признак
        log_scale (bool, optional): нужно ли использовать логорифмический масштаб Defaults to False.
        left (int, optional): левая граница итервала. Defaults to 3.
        right (int, optional): правая граница интервала. Defaults to 3.

    Returns:
        _type_: возвращает outliers - выбросы за границей интревала, cleaned - очищенные данные
    """
    if log_scale:
        x = np.log(data[feature]) # не прибавляем 0, так как у нас нет нулевых значений возраста
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    
    lower_bound = mu - left * sigma 
    upper_bound = mu + right * sigma
    
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x > lower_bound) & (x < upper_bound)]
    return outliers, cleaned
outliers, cleaned = outliers_z_score_mod(hh_data, 'Возраст', log_scale=1, right=4)
```

```python
print(f'Число выбросов по методу z-отклонения: {outliers.shape[0]}')
print(f'Число оставшихся значений по методу z-отклонения: {cleaned.shape[0]}')
display(outliers['Возраст'])
```

Распределение ассимитрично в правую сторону.
С учётом послабления правой границы на 1 получилось три выброса - 15, 15 и 100 лет
